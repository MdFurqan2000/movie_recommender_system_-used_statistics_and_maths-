```python
meanvote=df['vote_average'].mean()
print(meanvote)
```

```python
from numpy import minimum
minimumvote = df['vote_count'].quantile(0.81)
print(minimumvote)
```

```python
q_movies=df.copy().loc[df['vote_count']>=minimumvote]
q_movies.shape
```

```python
def weighted_rating(x,minimumvote=minimumvote,meanvote=meanvote):
    voters=x['vote_count']
    avg_vote=x['vote_average']
    ##calculating based on IMBD formula
    return(voters/(voters+minimumvote) * avg_vote) + (minimumvote/(minimumvote+voters) * meanvote)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score',ascending=False)
```
